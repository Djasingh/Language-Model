```python
import spacy
from spacy import displacy
```

<h2>Part Of Speech tagging using Spacy</h2>

```python
def pos_tagging(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    return doc
```

```python
input_str="Traditionally, we can use ngrams to generate language models to predict which word comes next given a history of words."
doc = pos_tagging(input_str)
for token in doc:
 print("text:{0}, lemma:{1}, pos:{2}, tag:{3}, dep:{4}, shape:{5}, is_alpha:{6}, is_stop:{7}".format(token.text, 
    token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop))
```

<h2> Dependency Parsing Using Spacy</h2>

```python
def dependency_parse(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    #displacy.serve(doc, style='dep')
    displacy.render(doc, style='dep')
    return doc
dependency_parse("Traditionally, we can use ngrams to generate language models to predict which word comes next given a history of words.")
```
